```python
ev_reg_counts = df['Model Year'].value_counts().sort_index()
ev_reg_counts
```

```python
from scipy.optimize import curve_fit
filtered_years = ev_reg_counts[ev_reg_counts.index <= 2023]
def exp_growth(x,a,b):
  return a*ny.exp(b*x)
x_data = filtered_years.index - filtered_years.index.min()
y_data = filtered_years.values
params, covariance = curve_fit(exp_growth, x_data, y_data)
forecast_years = ny.arange(2024,2024+6) - filtered_years.index.min()
forecast_values = exp_growth(forecast_years, *params)
forecast_evs = dict(zip(forecast_years + filtered_years.index.min(),forecast_values))
forecast_evs
```

```python
years = ny.arange(filtered_years.index.min(), 2029 + 1)
actual_years = filtered_years.index
forecast_years_full = ny.arange(2024,2029+1)
actual_values = filtered_years.values
forecast_values_full = [forecast_evs[years]for years in forecast_years_full]
plt.figure(figsize=(12,6))
plt.plot(actual_years, actual_values, 'bo-', label = 'Actual Registrations')
plt.plot(forecast_years_full, forecast_values_full, 'ro--', label = 'Forecast Registrations')
plt.title("Current & Estimated Electric Vehicle Registrations")
plt.xlabel("Year")
plt.ylabel("Number of Registrations")
plt.legend()
plt.tight_layout()
plt.show()
```
